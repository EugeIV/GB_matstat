#### Задача 1
Группа студентов изучает 10 различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть 4 разных занятия?

Определим число размещений из 10 дисциплин по 4 занятия в каждом.

```python
from math import factorial
```

```python
def arrangements(n, k):
    return int(factorial(n) / factorial(n - k))
```

```python
print("Вариантов составления рассписания: ")
print(arrangements(10, 4))
```

#### Задача 2
Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

Для решения этой задачи будем использовать формулу сочетаний

```python
from math import factorial
```

```python
def combination(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

Рассмотрим сначала ситуацию, когда выбирается один туз из четырех

```python
c1 = combination(4, 1)
```

Остальные три карты выбираются из 48 карт, тузы не рассматриваются, т.к. они уже выбраны.

```python
c2 = combination(48, 3)
```

В таком случае число сочетаний, когда из колоды в 52 карты, выбрано 4 карты, одна из которых туз, будет равно

```python
res = c1 * c2
print(res)
```
