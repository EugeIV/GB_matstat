Пусть X - дискретная случайная величина, равная числу правильных ответов. Она распределена по биномиальному закону с параметрами $n = 5$ (число вопросов), $p = \frac{1}{4}$(вероятность угадать ответ), поэтому вероятности вычисляются по формуле Бернулли:
$$ P(X=k)=C_n^k\cdot(1-p)^{n-k}\cdot p^k=C_5^k\cdot(\frac{3}{4})^{5-k}\cdot(\frac{1}{4})^k $$
Получаем:

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

#### Задача 2
Вероятность поймать рыбу при одном забрасывании удочки равна 0.01. Какова вероятность того, что число пойманных рыб при 200 забросах удочки будет не менее 5 и не более 10?

Вероятность при 200 забросах удочки поймать от 5 до 10 рыб будет равна сумме вероятностей по каждому параметру. Применим формулу Пуассона, в которой $m$ будет находится в интервале $[5, 10]$, а $\lambda=n\cdot p=200\cdot 0.01=2 $

```python
import numpy as np
from math import factorial

n = 200
p = 0.01
lambda_ = n * p
summa_p = 0
```

```python
for m in range(5, 11):
    p_m = (lambda_**m / np.math.factorial(m) * np.exp(-lambda_))
    summa_p += p_m
```

```python
print("Ответ на задачу =", summa_p)
```

#### Задача 3
В условиях предыдущей задачи сколько забросов удочки нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет поймана рыба?
Подсказка. Как посчитать P(X >= k) для величины X, имеющей биномиальное распределение? Здесь предстоит немножко покодить.

$$P(X>10) = 0.9$$
$$P(X>10) = 1 - P(X<=10)$$

```python
import numpy as np
def combinations(n, k):
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

```python
def bern(n, k, p):
    return combinations(n, k) * (p ** k) * ((1 - p) ** (n - k))
```

```python
def bern_obr(n, k, p):
    return (1 - sum(bern(n, i, p) for i in range(k)))
```

```python
p = 0.01
k = 10
n = 10
b = 0

while b < 0.9:
    n += 1
    b = bern_obr(n, k, p)
    
print(f"Нужно: {n} забросов, с вероятностью = {b}")
```
